# Preprocessing Garbase Data

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("lab_task.csv")   #import dataset which collected from Kaggle
```

```python
df.head()
```

```python
df.info()
```

# Drop unnecessary column from dataset

```python
df.drop('deck',axis=1, inplace= True)
df.drop('embarked',axis=1, inplace= True)
```

```python
df.head()
```

# Handle Missing Value

```python
df.isnull().sum()
```

```python
df['age']= df['age'].fillna(df['age'].mean())
```

```python
df2 = df.dropna(axis=0)
```

```python
df2.info()
```

# Discritization

```python
df2['Label'] = pd.cut(x=df2['age'], bins=[0, 3, 17, 63, 99],
labels=['Baby', 'Child', 'Adult','Elderly'])
```

```python
df2.head()
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
from pandas._libs.tslibs import dtypes
import warnings
warnings.filterwarnings('ignore')
for col in df2.columns:
  if df2[col].dtype == np.number:
    continue
  else:
     df2[col]=le.fit_transform(df2[col])
```

```python
df2.head()
```

# Normalization(Scalling dataset)

```python
from sklearn.preprocessing import MinMaxScaler
scal= MinMaxScaler()
```

```python
df2['fare'] = scal.fit_transform(df2[['fare']])
```
